```python
import numpy as np
import pandas as pd 
movie_dataset = pd.read_csv('preprocessed_movies.csv')
movie_dataset
```

```python
average_grades_dataset = pd.read_csv('calculated_average.txt', sep="\t", header=None)
average_grades_dataset.columns = ["movie_id", "average_grade"]
average_grades_dataset
```

```python
result = pd.merge(movie_dataset, average_grades_dataset, on="movie_id")
result.to_csv('movie_dataset_with_average_grades.csv', index=False)
```

```python
def show_best_rated_movies_by_director_name(parameter_value, dataset):

    series = result["director_name"]
    boolean_array = series.str.contains(parameter_value)
    
        
    requested_result = dataset[boolean_array]
    requested_result = requested_result.sort_values(by=['average_grade', 'movie_popularity'], ascending=False)
    requested_result = requested_result.reset_index()
    return requested_result
```

```python
def show_best_rated_movies_by_release_year(parameter_value, dataset):

    boolean_array = dataset["movie_release_year"] == parameter_value
        
    requested_result = dataset[boolean_array]
    
    requested_result = requested_result.sort_values(by=['average_grade', 'movie_popularity'], ascending=False)
    requested_result = requested_result.reset_index()
    return requested_result
```

```python
def show_best_similar_movies(movie_name, dataset):

    boolean_array = dataset["movie_name"] == movie_name
        
    requested_result = dataset[boolean_array]
    
    if len(requested_result) != 0:
        trimmed_similarity_data = requested_result["similarity_data"].replace({'"':''}, regex=True).replace({'\[':''}, regex=True).replace({'\]':''}, regex=True)
        similarity_data = trimmed_similarity_data.str.split(",", expand=True)

        data = {'movie_name': similarity_data[0], 'similarity_score': similarity_data[1], 'popularity': similarity_data[2]}

        formated_result = pd.DataFrame(data);    

        formated_result = formated_result.sort_values(by=['similarity_score', 'popularity'], ascending=False)
        formated_result = formated_result.reset_index()
        return formated_result
    return requested_result
```

```python
import numpy as np
import pandas as pd

def main():
    movies_with_average_grades_dataset = pd.read_csv('movie_dataset_with_average_grades.csv')
    movie_cosine_similarity_dataset = pd.read_csv('cloudProcessing.txt', sep="\t", header=None)
    movie_cosine_similarity_dataset.columns = ["movie_name", "similarity_data"]
    
    while (1==1):
        print("\n")
        print("Choose one of the options below: \n")
        print("1. year \n")
        print("2. dir\n")
        print("3. sim\n")
        print("4. break\n")
        parameter_type = input()
        if parameter_type == "year":
            print("_________________________________________________")
            print("\n")
            parameter_value = input("Type the release year: ")
            df = show_best_rated_movies_by_release_year(int(parameter_value), movies_with_average_grades_dataset)
            if len(df) == 0:
                print("\n")
                print("Unfortunatelly, there are no movies that match provided parameters")
                continue
            return df
        if parameter_type == "dir":
            print("_________________________________________________")
            print("\n")
            parameter_value = input("Type the director name: ")
            df = show_best_rated_movies_by_director_name(parameter_value, movies_with_average_grades_dataset)
            if len(df) == 0:
                print("\n")
                print("Unfortunatelly, there are no movies that match provided parameters")
                continue
            return df
        if parameter_type == "sim":
            print("_________________________________________________")
            print("\n")
            parameter_value = input("Type the movie name: ")
            df = show_best_similar_movies(parameter_value, movie_cosine_similarity_dataset)
            if len(df) == 0:
                print("\n")
                print("Unfortunatelly, there are no movies that match provided parameters")
                continue
            return df
        if parameter_type == "break":
            print("_________________________________________________")
            print("\n")
            print("Program shutdown")
            break
        else: 
            print("\n")
            print("_________________________________________________")
            print("\n")
            print("Please enter a valid filter")
            continue
```

```python
main()
```
